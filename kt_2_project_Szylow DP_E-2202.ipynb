<h3>Предобработка</h3>

<i><h4>Избавление от пропусков</h4>
<br>1. Посчитаем распределение безработных по полу и существующие пропуски заполним, сохраняя пропорцию</br></i>

```python
females = df[df['gender'] == 0.0]['id'].count()
females
```

```python
males = df[df['gender'] == 1.0]['id'].count()
males
```

```python
males_perc = round(males / (males + females), 2)
males_perc
```

<u>Таким образом, пропуски будут заполнены пополам</u>

```python
empty_indexes = list(df[df['gender'].isna()].index)
empty_indexes
```

```python
for i in range(len(empty_indexes)):
    if i % 2:
        df.loc[df.index[empty_indexes[i]], 'gender'] = 0.0
        print(df.iloc[empty_indexes[i]])
    else:
        df.loc[df.index[empty_indexes[i]], 'gender'] = 1.0
        print(df.iloc[empty_indexes[i]])
    i += 1
```

```python
df[df['gender'].isna()]
```

<i>2. Пропуски в образовании будем заполнять наиболее часто встречающимся значением</i>

```python
df[df['education'].isna()]
```

```python
df['education'].value_counts().sort_values(ascending=False)
```

```python
df['education'] = df['education'].fillna(df['education'].mode()[0])
df['education']
```

<h4>Структура трудоустройства по регионам внутри федерального округа</h4>

```python
def graphic_sixth_seventh_first(districtName):
    winners_regions = pd.DataFrame()
    winners_regions['region'] = data_2020.query('district == @districtName')['region'].unique()
    winners_regions[True] = data_2020.query('district == @districtName and find_new_job == True')['region'].value_counts().reset_index().sort_values(by='region')['count']
    winners_regions[False] = data_2020.query('district == @districtName and find_new_job == False')['region'].value_counts().reset_index().sort_values(by='region')['count']
    winners_regions['Процент нашедших работу'] = round(winners_regions[True] / (winners_regions[True] + winners_regions[False]) * 100, 2)
    winners_regions = winners_regions.sort_values(by='Процент нашедших работу')
    graphic = winners_regions.plot(kind='barh', x='region', y='Процент нашедших работу', title=f'Процент нашедших работу в {districtName}', legend=False)
    graphic = graphic.bar_label(graphic.containers[0])
    return graphic
```

```python
for i in data_2020['district'].unique():
    print(graphic_sixth_seventh_first(i))
```

<u>Выводы:</u>
    <i><br>1.Худшая ситуация с трудоустройством в Чечне - нашли работу менее 1% обратившихся</br>
    <br>2.Лучше всего трудоустройство в Ярославской области - 47%</br>
    <br><img src='https://special.pskov.ru/sites/default/files/documents/pskovskaya_oblast.png' width='40' align='left'>В Псковской области работу нашли 28,4% - хуже по СЗФО ситуация только в Республике Коми (26,4%), для сравнения в Новгородской области - 32,4%</br>
    <br>4. В СЗФО самое равномерное распределение трудоустройства, в пределах 29-30%</br>
